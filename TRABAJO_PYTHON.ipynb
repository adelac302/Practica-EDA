##ANALISIS DE GENERO

```python
# Agrupar por sexo y contar el número de registros
conteo= df.groupby('SEXO').count()['UUID']
# Calcular el porcentaje de datos
porcentaje = (conteo / len(df)) * 100
# Concatenar los conteos y los porcentajes
resultado = pd.concat([conteo, porcentaje], axis=1)
resultado.columns = ['Cantidad', 'Porcentaje']
print(resultado)
```

##ANALISIS DE FACULTAD Y ESCUELA

```python
df_fau_escu=df.groupby(['FACULTAD','ESCUELA_ACADEMICA'])
df_fe_agg=df_fau_escu.agg(
         cantidad2= pd.NamedAgg(column='UUID',aggfunc='count')
)
df_fe_agg
```

```python
# Preparar los datos
df_facultad2 = df_fe_agg.groupby('FACULTAD').sum()
# Calcular el porcentaje que representa cada facultad con respecto al total
porcentaje_facultades = (df_facultad2['cantidad2'] / df_facultad2['cantidad2'].sum()) * 100
# Definir una paleta de colores
palette = sns.color_palette("Set3", len(df_facultad2))

# Preparar los datos
sizes = df_facultad2['cantidad2'].values
labels = [f'{label}\n{porcentaje:.1f}%' for label, porcentaje in zip(df_facultad2.index, porcentaje_facultades)]

# Crear el treemap
plt.figure(figsize=(30, 10))
squarify.plot(sizes=sizes, label=labels, alpha=0.6)
plt.axis('off')  # Desactivar los ejes
plt.title('Treemap de Cantidad de Beneficiados  Porcentaje')
plt.show()
```

```python
# Preparar los datos
df_facultad3 = df_fe_agg.groupby('ESCUELA_ACADEMICA').sum()
# Seleccionar las 8 escuelas académicas con mayor cantidad de beneficiados
top = df_facultad3.nlargest(8, 'cantidad2')

# Crear el gráfico de barras
plt.figure(figsize=(12, 8))
plt.barh(top.index, top['cantidad2'], color='skyblue')
plt.xlabel('Cantidad de Beneficiados')
plt.ylabel('Escuela Académica')
plt.title('Top 8 Escuelas Académicas con Mayor Cantidad de Beneficiados')
plt.gca().invert_yaxis()  # Invertir el eje y para que la escuela con la mayor cantidad esté en la parte superior
plt.show()
```

###ANALISIS DE LA CATEGORIA ADOLESCENTE

```python
df_adol=df[df['EDAD_BINED']=='adolescente']
df_adol
```

```python
#Agrupamos por departamento
df_grouped=df_adol.groupby('DEPARTAMENTO')
#HACEMOS FUNCIONES DE AGRUPACIÓN DE LA DATA
agg_data=df_grouped.agg(
    cantidad=pd.NamedAgg(column='UUID',aggfunc='count'),
    edad_mean=pd.NamedAgg(column='EDAD', aggfunc='mean'),
    MASCULINO=pd.NamedAgg(column='SEXO',
                             aggfunc= lambda x :(x=='MASCULINO').sum()),
    FEMENINO=pd.NamedAgg(column='SEXO',
                            aggfunc=lambda y :(y=='FEMENINO').sum()),
)
agg_data
```

```python
# Seleccionar los tres primeros departamentos con mayor porcentaje
top_3_departamentos = agg_data.nlargest(3, 'cantidad')

# Extraer los valores de cantidad y los índices de agg_data para usarlos en el gráfico de pastel
valores = top_3_departamentos['cantidad'].values
indices = top_3_departamentos.index

# Graficar el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(valores, labels=indices, autopct='%1.1f%%')
plt.title('Top 3 Departamentos con Mayor Porcentaje de Beneficiados_Adolescentes')
plt.axis('equal')  # Asegurar que el gráfico de pastel sea un círculo
plt.show()
```

```python

# Lista de facultades de interés
facultades_interes = ['PSICOLOGÍA', 'ECONOMÍA', 'CIENCIAS SOCIALES', 'MEDICINA']
facultades_interes1 = ['PSICOLOGÍA', 'ECONOMÍA']
facultades_interes2 = ['CIENCIAS SOCIALES', 'MEDICINA']
facultades_interes3 = ['ECONOMÍA', 'CIENCIAS SOCIALES']

# Filtra los DataFrames según las facultades de interés
df_filtrado1 = df_fusionado1[df_fusionado1['Facultad'].isin(facultades_interes1)]
df_filtrado2 = df_fusionado2[df_fusionado2['Facultad'].isin(facultades_interes2)]
df_filtrado3 = df_fusionado3[df_fusionado3['Facultad'].isin(facultades_interes3)]

# Crear el gráfico
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.25

# Obtener las facultades presentes en cada DataFrame filtrado
facultades_presentes1 = set(df_filtrado1['Facultad'])
facultades_presentes2 = set(df_filtrado2['Facultad'])
facultades_presentes3 = set(df_filtrado3['Facultad'])

# Combinar las facultades presentes para obtener la lista de facultades únicas en los DataFrames filtrados
facultades_combinadas = list(facultades_presentes1 | facultades_presentes2 | facultades_presentes3)

# Crear un diccionario para mapear cada facultad a un índice
facultades_dict = {facultad: idx for idx, facultad in enumerate(facultades_combinadas)}

# Ajustar los DataFrames filtrados para contener solo las facultades combinadas
df_filtrado1 = df_filtrado1[df_filtrado1['Facultad'].isin(facultades_combinadas)]
df_filtrado2 = df_filtrado2[df_filtrado2['Facultad'].isin(facultades_combinadas)]
df_filtrado3 = df_filtrado3[df_filtrado3['Facultad'].isin(facultades_combinadas)]

# Calcular las posiciones de las barras para cada DataFrame filtrado
indices1 = df_filtrado1['Facultad'].map(facultades_dict)
indices2 = df_filtrado2['Facultad'].map(facultades_dict)
indices3 = df_filtrado3['Facultad'].map(facultades_dict)

# Graficar los datos de cada DataFrame filtrado
ax.barh(indices1, df_filtrado1['Raciones'], height=width, label='Desayuno', color='red')
ax.barh(indices2 + width, df_filtrado2['Raciones'], height=width, label='Almuerzo', color='green')
ax.barh(indices3 + 2 * width, df_filtrado3['Raciones'], height=width, label='Cena', color='blue')

# Configurar las etiquetas del eje y
ax.set_yticks([i + width for i in range(len(facultades_combinadas))])
ax.set_yticklabels(facultades_combinadas)

# Configurar etiquetas y título
ax.set_ylabel('Facultad')
ax.set_xlabel('Raciones')
ax.set_title('RANKING DE RACIONES EN GENERAL')

# Agregar la leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
